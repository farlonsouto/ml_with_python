```python
from scipy import sparse
# create an identity matrix
eye = np.eye(4)
print('NumPy array:\n', eye)
```

```python
# Convert the NumPy array to a SciPy sparse array in CSR format:
sparse_matrix = sparse.csr_matrix(eye)
print('\nSciPy sparse CSR matrix:\n', sparse_matrix)
```

```python
# Same identity matrix using the COO format:
data = np.ones(4)
row_i = np.arange(4)
col_j = np.arange(4)
eye_coo = sparse.coo_matrix((data, (row_i, col_j)))
print('COO sparse matrix coo:\n', eye_coo)
```

```python
# Matplot because one needs to visualize data:
# %matplotlib inline tells the notebook to print graphics inline, right beneath the cell
%matplotlib inline
import matplotlib.pyplot as plot

# generates a sequence of numbers from -7 to 9 with 50 steps in between
axis_x = np.linspace(-7, 9, 50)
# now an array with the sin of each value in the axis_x - This is the mindset one needs for using NumPy 
axis_y = np.sin(axis_x)
# Finally, we plot it in a simple way. That's why people use python.
# plot.plot(axis_x, axis_y, marker="o", ms=10, mfc='red', mec='black')
plot.plot(axis_x, axis_y, marker="x")
```
