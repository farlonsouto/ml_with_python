```python
# ==============================================================================
# COMPARISON: LINEAR REGRESSION vs. RIDGE vs. LASSO
# ==============================================================================
#
# 1. LINEAR REGRESSION (Ordinary Least Squares)
#    - The "Purest" form: Aims to minimize the sum of squared errors between 
#      predictions and actual data.
#    - Problem: It has no "leash." If features are correlated or noisy, 
#      coefficients can become massive, leading to terrible overfitting.
#
# 2. RIDGE REGRESSION (L2 Regularization)
#    - The "Shinker": Adds a penalty proportional to the SQUARE of the 
#      coefficients (controlled by Alpha).
#    - Outcome: It shrinks all coefficients toward zero but rarely makes them 
#      exactly zero. It keeps all features in the model but reduces their impact.
#    - Best used: When you have many features that all contribute a little bit.
#
# 3. LASSO REGRESSION (L1 Regularization)
#    - The "Selector": Adds a penalty proportional to the ABSOLUTE value of 
#      the coefficients.
#    - Outcome: It can force unimportant coefficients to become EXACTLY zero. 
#      This effectively performs automatic "Feature Selection."
#    - Best used: When you suspect only a few features are actually important 
#      and want a simpler, more interpretable model.
# ==============================================================================

from sklearn.linear_model import LinearRegression, Ridge, Lasso
from sklearn.model_selection import train_test_split
import mglearn
X, y = mglearn.datasets.load_extended_boston()
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)
lr = LinearRegression().fit(X_train, y_train)
print('Test score with Linear Regression : {:.2f}'.format(lr.score(X_test, y_test)))
# ==============================================================================
# ALPHA: THE REGULARIZATION PARAMETER
# ==============================================================================
# 1. PREVENTS OVERFITTING: Alpha's primary job is to keep the model from 
#    becoming too complex or "memorizing" noise in the training data.
#
# 2. PENALTY FOR COMPLEXITY: In standard Linear Regression, coefficients 
#    (slopes) can become huge and erratic. Alpha adds a "penalty" to the loss 
#    function based on the size of these coefficients. To keep the total 
#    "cost" low, the model is forced to keep coefficients small.
#
# 3. THE BALANCING ACT:
#    - Alpha = 0: Identical to standard Linear Regression. High risk of 
#      OVERFITTING (model is too complex).
#    - Alpha is Large: Strong penalty. Coefficients are pushed toward zero. 
#      High risk of UNDERFITTING (model is too simple/stable).
#
# 4. IMPROVING GENERALIZATION: By "shrinking" coefficients, Alpha reduces 
#    sensitivity to training noise, helping the model perform better on 
#    new, unseen data (the Test Set).
# ==============================================================================
for i in range(20):
    if i < 11:
        alpha_value = i/10
    else:
        alpha_value = i -10
    ridge = Ridge(alpha=alpha_value).fit(X_train, y_train)
    print('Test score Ridge with alpha {:.2f} : {:.2f}'.format(alpha_value, ridge.score(X_test, y_test)))
```

```python
import numpy as np
for alpha_value in [0.1, 0.001, 1]:
    lasso = Lasso(alpha=alpha_value, max_iter=100000).fit(X_train, y_train)
    print('Test score Lasso with alpha {:.2f} : {:.2f}'.format(alpha_value, lasso.score(X_test, y_test)))
    print('Number of features used :', np.sum(lasso.coef_ != 0))
```
